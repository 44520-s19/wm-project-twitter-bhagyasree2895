```python
def lexical_diversity(text):
    tokens=text.lower().split()
    return len(set(tokens))/len(tokens)


# Samantha Diversities
#print(lexical_diversity(samantha_tweets[0].text))
for i in range(len(samantha_tweets)):
    lexical_diversity(samantha_tweets[i].text)
#print(len(samantha_tweets))

sam_diversities=[]
for tweet in samantha_tweets:
    sam_diversities.append(lexical_diversity(tweet.text))

print(sam_diversities)
print(min(sam_diversities))

```

```python
# Kajal Diversities
for i in range(len(kajal_tweets)):
    lexical_diversity(kajal_tweets[i].text)
#print(len(kajal_tweets))

kajal_diversities=[]
for tweet in kajal_tweets:
    kajal_diversities.append(lexical_diversity(tweet.text))

print(kajal_diversities)
print(min(kajal_diversities))
```

```python
# Rashmika Diversities
for i in range(len(rashmika_tweets)):
    lexical_diversity(rashmika_tweets[i].text)
#print(len(rashmika_tweets))

rashmika_diversities=[]
for tweet in rashmika_tweets:
    rashmika_diversities.append(lexical_diversity(tweet.text))

print(rashmika_diversities)
print(min(rashmika_diversities))
```

```python
import numpy as np
import matplotlib.pyplot as plt

xax=sam_diversities
yax=samantha_retweet

xs=kajal_diversities
ys=kajal_retweet

xx=rashmika_diversities
yy=rashmika_retweet


plt.scatter(xs, ys,c='green',marker=r'$\sigma$',label='Samantha')
plt.scatter(xax,yax,c='r',marker=r'$\clubsuit$',label='Kajal Agarwal')
plt.scatter(xx,yy,c='b',marker='+',label='Rashmika')

plt.xlabel('Diversities')
plt.ylabel('Retweets')


print(' ')
plt.suptitle('Analysis of Tweets on Lexical Diversities and Their Retweets', fontsize=14, fontweight='bold',color='green')
plt.legend(loc='best',shadow=True)
plt.show()


#sentiments=[get_sentiments(tweet.text) for tweet in tweets]

#build our xs, ys, and zs
```
