### Lexical diversity: 
- "the range of different words used in a text, with a greater range indicating a higher diversity"
- If a text has a higher index of LD it is likely to be more complex, more advanced and more difficult 
- One of the most commonly used approaches to measure LD is to use the ratio of unique lexical items divided by the total number of words in a sample after standardizing the length of the sample.
- In computational linguistics, lexical richness is a measure of how many tokens (individual words and punctuation) occur in a given text, divided by how many types (unique words and punctuation) occur in that same text.
- The higher the number, the more lexically rich it is, meaning, the more unique words (and punctuation) it contains compared to the total number of words and punctuation it contains.

### Description of analysis steps: 

1) Calculating lexical diversity for tweets, wrote a function for computing lexical diversity
- Send any text as in input
- Then Convert the input text into lowercase using lower() in built method for strings
- Then take this lowercase text and Split the given text using split() in built method for strings
- Calculate the length of splitted text and Calculate the length of set(splitted text) which stores only unique words
- Calculate the ratio of len(unique words in that given input text) to len(words in that given input text)
- This ratio should be in range of 1(i.e 0 to 1) which is known as Lexical diversity
    
2) Apply the above lexical_diversity() function for all the tweets of Samantha using for loop 

3) And append those values to a list called sam_diversities

4) Apply the above lexical_diversity() function for all the tweets of Kajal using for loop

5) And store those values in a list called kajal_diversities

6) Apply the above lexical_diversity() function for all the tweets of Rashmika using for loop

7) And store those values in a list called rashmika_diversities

8) Calculate number of retweets for each tweet of samantha and store those values in samantha_retweet list

9) Calculate number of retweets for each tweet of Kajal and store those values in kajal_retweet list

10) Calculate number of retweets for each tweet of Rashmika and store those values in rashmika_retweet list

11) We will plot a graph, After finding Lexical diveristies and number of retweets of 3 different persons tweets

12) Considering X-axis as Diversities and Y-axis as Number of Retweets

13) Scatter the points as (lexical diversity, retweets) for 3 different persons

14) After scattering all points, conclude who is more balancing with lexical diverity and number of retweets.


```python
# Analysis (code)

def lexical_diversity(text):
    tokens=text.lower().split()
    return len(set(tokens))/len(tokens)

#Analyzing data based on Lexical diversities and Number of Retweets
#print("************ Tweets ***************")


# Samantha Diversities

sam_diversities=[]
for tweet in samantha_tweets:
    sam_diversities.append(lexical_diversity(tweet.text))

#print(sam_diversities)
#print(min(sam_diversities))


# Kajal Diversities

kajal_diversities=[]
for tweet in kajal_tweets:
    kajal_diversities.append(lexical_diversity(tweet.text))

#print(kajal_diversities)
#print(min(kajal_diversities))



# Rashmika Diversities


rashmika_diversities=[]
for tweet in rashmika_tweets:
    rashmika_diversities.append(lexical_diversity(tweet.text))

#print(rashmika_diversities)
#print(min(rashmika_diversities))



# Retweets

samantha_retweet=[]
for i in range(len(samantha_tweets)):
    samantha_retweet.append(len(samantha_tweets[i].text.split()))
#print(samantha_retweet)



kajal_retweet=[]
for i in range(len(kajal_tweets)):
    kajal_retweet.append(len(kajal_tweets[i].text.split()))
#print(kajal_retweet)



rashmika_retweet=[]
for i in range(len(rashmika_tweets)):
    rashmika_retweet.append(len(rashmika_tweets[i].text.split()))
#print(rashmika_retweet)
```

```python
# Presentation of results (code)


#Display/plot your data using matplotlib or other python mechanism (note: this should be done live so that if your data changes so does the visualization)


import numpy as np
import matplotlib.pyplot as plt

xax=sam_diversities
yax=samantha_retweet

xs=kajal_diversities
ys=kajal_retweet

xx=rashmika_diversities
yy=rashmika_retweet


plt.scatter(xs, ys,c='green',marker=r'$\sigma$',label='Samantha')
plt.scatter(xax,yax,c='r',marker=r'$\clubsuit$',label='Kajal Agarwal')
plt.scatter(xx,yy,c='b',marker='+',label='Rashmika')

plt.xlabel('Diversities')
plt.ylabel('Retweets')


print(' ')
plt.suptitle('Analysis of Tweets on Lexical Diversities and Their Retweets', fontsize=14, fontweight='bold',color='green')
plt.legend(loc='best',shadow=True)
plt.show()
```

```python
# Drawing/Presenting more appropriately using Barplots using errorbars

# Calculate the average
samantha_mean = np.mean(samantha_retweet)
kajal_mean = np.mean(kajal_retweet)
rashmika_mean = np.mean(rashmika_retweet)
print('*************Retweets*************')
print('Samantha',samantha_mean)
print('Kajal',kajal_mean)
print('Rashmika',rashmika_mean)

# Calculate the standard deviation
samantha_std = np.std(samantha_retweet)
kajal_std = np.std(kajal_retweet)
rashmika_std = np.std(rashmika_retweet)
# Create lists for the plot
materials = ['Samantha', 'Kajal', 'Rashmika']
x_pos = np.arange(len(materials))
CTEs = [samantha_mean, kajal_mean, rashmika_mean]
error = [samantha_std, kajal_std, rashmika_std]



# Build the plot
fig, ax = plt.subplots()
ax.bar(x_pos, CTEs, yerr=error, align='center', alpha=0.5, ecolor='black', capsize=10)
ax.set_ylabel('Coefficient of Retweets')
ax.set_xticks(x_pos)
ax.set_xticklabels(materials)
ax.set_title('Retweets Three heroines',fontsize=14, fontweight='bold',color='green')
ax.yaxis.grid(True)

# Save the figure and show
plt.tight_layout()
#plt.savefig('bar_plot_with_error_bars.png')
plt.show()


# Calculate the average
samantha_mean = np.mean(sam_diversities)
kajal_mean = np.mean(kajal_diversities)
rashmika_mean = np.mean(rashmika_diversities)


print( )
print( )
print( )
print('*************Lexical diversities*************')
print('Samantha',samantha_mean)
print('Kajal',kajal_mean)
print('Rashmika',rashmika_mean)

# Calculate the standard deviation
samantha_std = np.std(sam_diversities)
kajal_std = np.std(kajal_diversities)
rashmika_std = np.std(rashmika_diversities)
# Create lists for the plot
materials = ['Samantha', 'Kajal', 'Rashmika']
x_pos = np.arange(len(materials))
CTEs = [samantha_mean, kajal_mean, rashmika_mean]
error = [samantha_std, kajal_std, rashmika_std]



# Build the plot
fig, ax = plt.subplots()
ax.bar(x_pos, CTEs, yerr=error, align='center', alpha=0.5, ecolor='black', capsize=10)
ax.set_ylabel('Coefficient of Lexical diversities')
ax.set_xticks(x_pos)
ax.set_xticklabels(materials)
ax.set_title('Lexical diversities of Three heroines',fontsize=14, fontweight='bold',color='green')
ax.yaxis.grid(True)

# Save the figure and show
plt.tight_layout()
#plt.savefig('bar_plot_with_error_bars.png')
plt.show()
```
